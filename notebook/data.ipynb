```python
import pandas as pd
import numpy as np
import regex as re
```

```python
bfilms_json = pd.read_json('../data/BFilms_db.json')
bfilms_db = pd.DataFrame(bfilms_json)
```

```python
bfilms_db = bfilms_db.T
```

```python
bfilms_db.head()
```

```python
def movielength(value):
    try:
        hours = re.search(r'(\d+)h', value)
        mins = re.search(r'(\d+)m', value)
        
        return int(hours.group(1)) * 60 + int(mins.group(1))
    
    except:
        
        try:
            hours = re.search(r'(\d+)h', value)
    
            return int(hours.group(1)) * 60 
        
        except:
            
            try:
                mins = re.search(r'(\d+)m', value)
        
                return int(mins.group(1))
            
            except:
                pass
```

```python
bfilms_db['Length'] = bfilms_db['Length'].apply(movielength)
```

```python
bfilms_db.head()
```

```python
def get_odd_cast_members(cast_string):
    cast_list = cast_string.split(',')
    odd_cast_members = [cast_list[i] for i in range(len(cast_list)) if i % 2 == 0]
    return odd_cast_members
```

```python
bfilms_db['Cast'] = bfilms_db['Cast'].apply(get_odd_cast_members)
```

```python
bfilms_db.head()
```

```python
bfilms_db_2 = bfilms_db.copy()
```

```python
cast_dict = {}
for i, cast_member in enumerate(bfilms_db_2['Cast'].explode().unique()):
    cast_dict[cast_member] = i + 1

cast_dict
```

```python
# Define a function to replace cast member names with IDs
def replace_cast_names(cast_lst):

    cast_ids = [str(cast_dict[cast_member]) for cast_member in cast_lst]
    
    return cast_ids
```

```python
# Apply the function to the 'Cast' column and store the result in a new column 'Cast IDs'
bfilms_db_2['Cast IDs'] = bfilms_db_2['Cast'].apply(replace_cast_names)
```

```python
bfilms_db_2['film_id'] = [i for i in range(1,len(bfilms_db_2)+ 1)]
```

```python
def get_genre(genre_string):
    genre_list = genre_string.split(',')
    genre = [genre_list[i] for i in range(len(genre_list))]
    return genre
```
